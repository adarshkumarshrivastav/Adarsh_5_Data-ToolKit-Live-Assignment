```python
#Q.3:- Explain the following terms:

The difference in np.array, np.asarray and np.asanyarray. The difference between Deep copy and shallow copy.
#Ans:-
####    np.array:

This function creates a new NumPy array object, either from a sequence-like object (like lists or tuples) or from an existing array.
If the input is already a NumPy array, np.array will create a copy of the array by default unless copy=False is explicitly passed. This ensures that any changes made to the original array will not affect the newly created array.
 Example
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array(arr1)
arr2[0] = 99
print(arr1)  # [1, 2, 3] - arr1 is unchanged

##  np.asarray:

This function converts the input to a NumPy array, but if the input is already a NumPy array, it does not copy the array. Instead, it returns the original array.
This is useful when you want to ensure that the input is an array but avoid the overhead of copying data if it’s already an array.
Example:

arr1 = np.array([1, 2, 3])
arr2 = np.asarray(arr1)
arr2[0] = 99
print(arr1)  # [99, 2, 3] - arr1 is modified because arr2 is just a reference
 
###    np.asanyarray:

Similar to np.asarray, but with one key difference: if the input is a subclass of a NumPy array (like a masked array), it will not convert it to a base ndarray. Instead, it will return the subclass object itself.
This is useful when you want to maintain the type of the input array (e.g., if it’s a masked array) rather than converting it to a standard NumPy array.
Example:


arr1 = np.array([1, 2, 3])
arr2 = np.asanyarray(arr1)
print(type(arr2))  # <class 'numpy.ndarray'>
masked_arr = np.ma.array([1, 2, 3], mask=[0, 1, 0])
arr3 = np.asanyarray(masked_arr)
print(type(arr3))  # <class ''numpy.ma.core.MaskedArray'>
2. Difference Between Deep Copy and Shallow Copy

Shallow Copy:

A shallow copy of an object creates a new object, but does not create copies of nested objects within the original object. Instead, it copies references to the nested objects, so changes to the nested objects in the copy will also affect the original object.
In Python, this can be done using the copy() method or the copy module’s copy() function.
Example:

import copy
original = [[1, 2, 3], [4, 5, 6]]
shallow_copy = copy.copy(original)
shallow_copy[0][0] = 99
print(original)  # [[99, 2, 3], [4, 5, 6]] - original is modified


Deep Copy:

A deep copy of an object creates a new object, and recursively copies all objects nested within the original object. This means the copy is completely independent of the original, and changes to the copy do not affect the original.
In Python, this can be done using the copy module’s deepcopy() function.
Example:

import copy
original = [[1, 2, 3], [4, 5, 6]]
deep_copy = copy.deepcopy(original)
deep_copy[0][0] = 99
print(original)  # [[1, 2, 3], [4, 5, 6]] - original is unchanged
```

```python
#Q.5:-Create a NumPy array with randorn integers between 1 and 10 of shape (5, 6). After creating the array perform the following operations

#a) Extract all even integers from array.

#b) Extract all add integers from array.
#Ans:-
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
array = np.random.randint(1, 11, size=(5, 6))

# Extract all even integers from the array
even_integers = array[array % 2 == 0]

# Extract all odd integers from the array
odd_integers = array[array % 2 != 0]

print("Original Array:\n", array)
print("\nEven Integers:\n", even_integers)
print("\nOdd Integers:\n", odd_integers)
```

```python
#Q.7:- Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column.
#Ans:-
import pandas as pd 
import numpy as np
df = pd.read_csv('People Data.csv')

def clean_phone(phone):
    if pd.isna(phone) or phone == '':
        return np.nan
  Remove non-numeric characters
    cleaned_phone = ''.join(filter(str.isdigit, phone))
     Return NaN if cleaned_phone is empty, otherwise return the cleaned phone number
    return np.nan if cleaned_phone == '' else int(cleaned_phone)

# Apply the cleaning function to the 'Phone' column
df['Phone'] = df['Phone'].apply(clean_phone)

# Convert the 'Phone' column to Int64 to handle integers with NaNs
df['Phone'] = df['Phone'].astype('Int64')

# Configure pandas to display numbers in full without scientific notation
pd.set_option('display.float_format', '{:.0f}'.format)

# Display the DataFrame and its data types
print("DATAFRAME : \n",df)
print("\nData Types :\n",df.dtypes)
```

```python
#18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
#Wave Function'

from bokeh.plotting import figure, show

x = list(range(100))

y = [np.sin(i) for i in x]

p = figure(title="Sine Wave Function", y_range=(-1.1, 1.1))

p.line(x, y, line_width=2)

p.grid.grid_line_color = None

show(p)
```

```python
#19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

from bokeh.plotting import figure, show
from bokeh.palettes import Spectral8

data = {
    "categories": list("ABCDEFGH"),
    "values": np.random.randint(low=1, high=100, size=8),
    "colors": Spectral8
}

p = figure(title="Bar Chart", y_range=(0, data["values"].max() + 10))

p.vbar(x="categories", top="values", width=0.5, color="colors", source=data)

show(p)
```

```python
#20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

import plotly.graph_objects as go

x = list(range(100))
y = [np.random.randint(1, 100) for _ in range(100)]

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines'))

fig.update_layout(
    title="Simple Line Plot",
    xaxis_title="X Axis",
    yaxis_title="Y Axis",
)

fig.show()
```
